```python
import torch
import os
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration

# Directory to cache the model
model_dir = "./blip1_model"

# Always load the processor from the pretrained repository
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")

# Check if the model is cached locally; if not, download and save it
if not os.path.exists(model_dir):
    os.makedirs(model_dir)
    model = BlipForConditionalGeneration.from_pretrained(
        "Salesforce/blip-image-captioning-base"
    )
    model.save_pretrained(model_dir)
    print("Model saved locally.")
else:
    model = BlipForConditionalGeneration.from_pretrained(model_dir)
    print("Model loaded from local storage.")

# Use GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Function to generate captions
def generate_caption(image_path):
    image = Image.open(image_path).convert("RGB")
    inputs = processor(images=image, return_tensors="pt").to(device)

    with torch.no_grad():
        output = model.generate(**inputs, max_new_tokens=50)

    caption = processor.tokenizer.decode(output[0], skip_special_tokens=True)
    return caption


```

```python
image_path = "/content/image.jpg"  # Replace with your image path
caption = generate_caption(image_path)
print("Generated Caption:", caption)
```
